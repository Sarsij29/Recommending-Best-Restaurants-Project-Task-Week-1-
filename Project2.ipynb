```python
# import pandas lib as pd
import pandas as pd
 
# read by default 1st sheet of an excel file
df = pd.read_excel('Copy of data.xlsx')
print(type(df))
df.head()
df_new = pd.read_excel('Country-Code.xlsx')
# df_new.head()
df2 = pd.merge(df,df_new,on='Country Code',how='left')
df2.head()
df2.isnull().sum() 
df2[df2['Restaurant Name'].isnull()]
```

```python
df2.dropna(axis=0,subset=['Restaurant Name'],inplace=True)
df2.reset_index(drop=True,inplace=True)
df2[df2['Cuisines'].isnull()]
```

```python
df2['Cuisines'].fillna('Others',inplace=True)
df2.isnull().sum()
df2.info()
```

```python
country_rest = df2.groupby(['Country Code','Country']).agg( Count = ('Restaurant ID','count'))
print(country_rest.sort_values(by='Count',ascending=False))
country_rest.plot(kind='barh')
country_rest = df2.groupby(['Country','City']).agg(Count = ('Restaurant ID','count'))
print(country_rest.describe())
print(country_rest.sort_values(by='Count',ascending=False))

min_rest = country_rest[country_rest['Count']==1]
min_rest.info()
min_rest
```

```python
import math
Rest_delivery=df2['Has Table booking'].to_list()
countYes=0
countNo=0
for i in range(len(Rest_delivery)):
  if Rest_delivery[i]=='Yes':countYes+=1
  else: countNo+=1
# print(countYes,countNo)
x=math.gcd(countYes,countNo)
print("ratio between restaurants that allow table booking vs. those that do not allow table booking --> "+str(countYes//x)+":"+str(countNo//x))
countOnline=0
for i in df2['Has Online delivery'].to_list():
  if i=='Yes':countOnline+=1
print("percentage of restaurants providing online delivery --> "+str((countOnline/len(df2['Has Online delivery'].to_list()))*100))
od=df2['Has Online delivery'].to_list()
vote=df2['Votes'].to_list()
# print(len(od)==len(vote))
vy=0
vn=0
for i in range(len(od)):
  if od[i]=='Yes':
    vy+=int(vote[i])
  else:vn+=int(vote[i])
print("difference in number of votes for the restaurants that deliver and the restaurants that do not deliver --> "+str(abs(vy-vn)))
```
